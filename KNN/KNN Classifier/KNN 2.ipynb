```python
from sklearn.datasets import load_breast_cancer
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=load_breast_cancer()
dataset
```

```python
#to display all columns
pd.pandas.set_option('display.max_columns',None)

df_cancer=pd.DataFrame(dataset['data'],columns=dataset['feature_names'])
df_cancer.head()
```

```python
df_cancer['type']=dataset['target']
df_cancer.head()
```

```python
dataset.target_names
```

```python
df_cancer.isnull().sum()
```

```python
df_cancer.isna().sum() #isna means 'is not available'
```

```python
df_cancer.shape
```

```python
x=df_cancer.iloc[:,0:30]
y=df_cancer.loc[:,'type']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
print('shape of xtrain data: ',X_train.shape)
print('shape of xtest data: ',X_test.shape)
print('shape of ytrain data: ',y_train.shape)
print('shape of ytest data: ',y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=7)
```

```python
model.fit(X_train,y_train)
```

```python
prediction=model.predict(X_test)
prediction
```

```python
# to know how much 'predicted values' are accurate we use accuracy

from sklearn.metrics import accuracy_score,confusion_matrix
accuracy=accuracy_score(y_test,prediction)
print('KNN model accuracy is:',accuracy)
```

```python
#if  want to check the correct prediction of zero class and one class
#use confusion matrix
```

```python
confusion_matrix(y_test,prediction) #diagonal elements are correct classification,rest all are incorrect classification
```

```python
#160-classify correctly(55 and 105)
160/171 # correct prediction score
#11-classify incorrectly(8,3)
```

## To check for which value of 'k' ,the maximum accuracy will come ?

```python
#1. use K and accuracy graph,(to check at which value of k the accuracy will maximum)
acc=[]
for i in range(1,41):
    model=KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train,y_train)
    pv=model.predict(X_test)
    acc.append(accuracy_score(y_test,pv))
    
plt.plot(range(1,41),acc,ls='--',marker='s',markersize=5,color='r',mfc='blue')
plt.xlabel('Number of Neighbours')
plt.ylabel('Accuracy')
plt.title('K and accuracy graph')

#by looking graph we can say that at k=7,the accuracy is maximum
```

```python
max(acc)
```

```python
acc.index(max(acc))
```

```python
# second we can check error rate by plotting 

error_rate=[]
for i in range(1,41):
    model=KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train,y_train)
    pv=model.predict(X_test)
    error_rate.append(np.mean(pv!=y_test))
    
plt.plot(range(1,41),error_rate,ls='--',marker='s',markersize=5,color='r',mfc='blue')
plt.xlabel('Number of Neighbours')
plt.ylabel('Error rate')
plt.title('K and error rate graph')

#by seeing the graph we can say the the error rate is minimum at k=7
```

```python
# at what test_size value we take to get maximum accuracy ?
acc=[]
for i in np.arange(0.10,1.0,0.10):
    #print(i)
    X_train, X_test, y_train, y_test=train_test_split(x,y,test_size=i,random_state=1)
    model=KNeighborsClassifier(n_neighbors=8)
    model.fit(X_train,y_train)
    pv=model.predict(X_test)
    acc.append(accuracy_score(y_test,pv))

plt.bar(np.arange(0.10,1,0.10),acc,width=0.08)
plt.xticks(np.arange(0.10,1,0.10))
plt.xlabel('test size')
plt.ylabel('accuracy')
plt.title('Test size and accuracy graph')
    
```
