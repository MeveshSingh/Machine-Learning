```python
from sklearn.datasets import load_wine
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset=load_wine()
dataset
```

```python
dataset.feature_names
```

```python
dataset.target_names
```

```python
df_wine=pd.DataFrame(dataset['data'],columns=dataset['feature_names'])
df_wine
```

```python
df_wine['wine class']=dataset['target']
df_wine
```

```python
#adding extra column and putting 'class' word in target 
df_wine['wine name']=df_wine['wine class'].replace({0:'class_0',1:'class_1',2:'class_2'})
df_wine
```

```python
X=df_wine.loc[:,'alcohol':'proline']
Y=df_wine['wine class']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.30,random_state=1)
print('shape of xtrain:',xtrain.shape)
print('shape of xtest:',xtest.shape)
print('shape of ytrain:',ytrain.shape)
print('shape of ytest:',ytest.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(xtrain,ytrain)
```

```python
pv=model.predict(xtest)
pv
```

```python
from sklearn.metrics import accuracy_score
accu=accuracy_score(ytest,pv)
print(f'Accuracy of the model is: {accu}')
```

```python
#feature importances
imp=model.feature_importances_
print(imp)
print()      
name=dataset.feature_names
print(name)
```

```python
plt.figure(figsize=(25,10))
plt.bar(name,imp)
```

```python
#based on feature importances:
#So now we are creating a new dataframe for important features

df_new_wine=pd.DataFrame(df_wine[['malic_acid','flavanoids','nonflavanoid_phenols','color_intensity','od280/od315_of_diluted_wines','proline']])
```

```python
df_new_wine
```

```python
y=dataset['target']
y
```

```python
from sklearn.model_selection import train_test_split
xtrain1,xtest1,ytrain1,ytest1=train_test_split(df_new_wine.loc[:,:],y,test_size=0.30,random_state=1)
print(f'shape of xtrain1:',xtrain1.shape)
print(f'shape of xtest1:',xtest1.shape)
print(f'shape of ytrain1:',ytrain1.shape)
print(f'shape of ytest1:',ytest1.shape)
```

```python
model2=DecisionTreeClassifier()
model2.fit(xtrain1,ytrain1)
```

```python
pv2=model2.predict(xtest1)
```

```python
accu2=accuracy_score(ytest1,pv2)
print('accuracy for model2: ',accu2)
```

```python
#here we are printing the accuracy of both models
print(f'Accuracy of model_1 is: {accu}')
print(f'Accuracy of model_2 is: {accu2}')
```

```python
#other information we get from (model.tree_.)

child_l=model2.tree_.children_left
child_r=model2.tree_.children_right
feat1=model.tree_.feature #for model 1
feat2=model2.tree_.feature #for model 2
threshold=model2.tree_.threshold

#printing all the above features

print(f'right children nodes: {child_r}')
print(f'left children nodes: {child_l}')
print(f'features used in old model: {feat1}')
print(f'features usesd in new model: {feat2}')
print(f'threshold of model: {threshold}')
```

## pydotplus ---for colored visualization of tree

```python
import pydotplus
from sklearn import tree
import collections
```

```python
data=tree.export_graphviz(model3,feature_names=df_new_wine.columns.to_list(), class_names=c_names, filled=True, rounded=True)
graph= pydotplus.graph_from_dot_data(data)
colors=('orange','blue', 'torquois')
edges=collections.defaultdict(list)
```

```python
for edge in graph.get_edge_list():
    edges[edge.get_source()].append(int(edge.get_destination()))
```

```python
edges
```

```python
for edge in edges:
    edges[edge].sort()
    for i in range(2):
        dest=graph.get_node(str(edges[edge][i]))[0]
        dest.set_fillcolor(colors[i])
graph.write_png('color_tree.png')

#Try this in Google Colab
```
