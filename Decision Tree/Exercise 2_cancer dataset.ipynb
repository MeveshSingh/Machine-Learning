```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_breast_cancer
```

```python
dataset=load_breast_cancer()
dataset
```

```python
df_cancer=pd.DataFrame(dataset['data'],columns=dataset['feature_names'])
df_cancer
```

```python
df_cancer.columns.nunique()
```

```python
df_cancer['result']=dataset['target']
df_cancer
```

```python
df_cancer['result name']=df_cancer['result'].replace({0:'malignant',1:'benign'})
df_cancer
```

```python
X=df_cancer.loc[:,'mean radius':'worst fractal dimension']
Y=df_cancer['result']
Y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.30,random_state=1)
print(f'shape of xtrain: {xtrain.shape}')
print(f'shape of xtest: {xtest.shape}')
print(f'shape of ytrain: {ytrain.shape}')
print(f'shape of ytest: {ytest.shape}')
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(xtrain,ytrain)
```

```python
pv=model.predict(xtest)
pv
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accu=accuracy_score(ytest,pv)
accu
```

```python
cm=confusion_matrix(ytest,pv)
cm
```

```python
#feature importances
imp=model.feature_importances_
print(imp)
print()      
name=dataset.feature_names
print(name)
```

```python
plt.figure(figsize=(60,10))
plt.bar(name,imp)
```
